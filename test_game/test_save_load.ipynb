# Graphs for tba

```python
import numpy as np
import pandas as pd
from pandas.io.parsers import read_csv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = read_csv('experimental_output.csv')
df
```

```python
plt.figure(figsize=(9.6, 7.2))
plt.scatter(df['size'], df['simple'])
plt.scatter(df['size'], df['binary'])
plt.scatter(df['size'], df['json'])
plt.scatter(df['size'], df['xml'])
plt.xlim(left=1)
plt.ylim(bottom=1)
plt.plot()

plt.xlabel("size N")
plt.ylabel("total time to save/load ($\mu$s)")
plt.legend()
plt.savefig('total_times_nonlog.png')
```

```python
plt.figure(figsize=(9.6, 7.2))
plt.scatter(df['size'], df['simple'])
plt.scatter(df['size'], df['binary'])
plt.scatter(df['size'], df['json'])
plt.xlim(left=1)
plt.ylim(bottom=1)
plt.plot()

plt.xlabel("size N")
plt.ylabel("total time to save/load ($\mu$s)")
plt.legend()
plt.savefig('total_times_noxml.png')
```

```python
plt.figure(figsize=(9.6, 7.2))
plt.scatter(df['size'], df['simple'])
plt.scatter(df['size'], df['binary'])
plt.scatter(df['size'], df['json'])
plt.scatter(df['size'], df['xml'])
plt.xscale('log')
plt.yscale('log')
plt.xlim(left=1)
plt.ylim(bottom=1)
plt.plot()

plt.xlabel("size N")
plt.ylabel("total time to save/load ($\mu$s)")
plt.legend()
plt.savefig('total_times_log.png')
```

```python
plt.figure(figsize=(9.6, 7.2))
plt.scatter(df['size'], df['simple_save'])
plt.scatter(df['size'], df['binary_save'])
plt.scatter(df['size'], df['json_save'])
plt.scatter(df['size'], df['xml_save'])
plt.xscale('log')
plt.yscale('log')
plt.xlim(left=1)
plt.ylim(bottom=1)
plt.plot()

plt.xlabel("size N")
plt.ylabel("time to save ($\mu$s)")
plt.legend()
plt.savefig('save_times.png')
```

```python
plt.figure(figsize=(9.6, 7.2))
plt.scatter(df['size'], df['simple_load'])
plt.scatter(df['size'], df['binary_load'])
plt.scatter(df['size'], df['json_load'])
plt.scatter(df['size'], df['xml_load'])
plt.xscale('log')
plt.yscale('log')
plt.xlim(left=1)
plt.ylim(bottom=1)
plt.plot()

plt.xlabel("size N")
plt.ylabel("time to load ($\mu$s)")
plt.legend()
plt.savefig('load_times.png')
```
